```python
#import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.svm import SVR
from sklearn import metrics
```

```python
#import data
df = pd.DataFrame(pd.read_csv(r'C:\Users\ibrah\OneDrive\Desktop\Intro to ML\HW4\Housing.csv'))
df.head() # To get first n rows from the dataset default value of n is 5
#m = len(df)
```

```python
#pre-process data - yes & no to binary
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

#delete furnished lined
del df['furnishingstatus']

#define map function
def binary_map(x):
    return x.map({'yes': 1, "no": 0})

#Apply fucntion
df[varlist] = df[varlist].apply(binary_map)
```

```python
df.head()
```

```python
#scale the set
df = StandardScaler().fit_transform(df) 
#Split Data set into Training Data and Test Data.
#np.random.seed(0)
#df_train,df_test = train_test_split(df, train_size=0.8, test_size=0.2)
#mT = len(df_train)
#mV = len(df_test)
```

```python
Y = df[:, 1]
X = df[:, 2:12]

Xd = pd.DataFrame(X)
```

```python
for k in range(1, 11):
    #Apply pca
    pca = PCA(n_components=k) 
    principalComponents = pca.fit_transform(Xd) 
    principalDf = pd.DataFrame(data = principalComponents) 
    
    #split set into training(0.8) and validations(0.2) sets
    X_train, X_test, Y_train, Y_test = train_test_split(principalDf, Y, test_size = 0.2)
    
    svr_rbf = SVR(kernel='rbf', C=1, gamma=0.1)
    svr_lin = SVR(kernel='linear', C=1)
    svr_poly = SVR(kernel='poly', C=1)
    
    y_rbf = svr_rbf.fit(X_train, Y_train).predict(X_test)
    y_lin = svr_lin.fit(X_train, Y_train).predict(X_test)
    y_poly = svr_poly.fit(X_train, Y_train).predict(X_test)
    
    MSE_rbf = metrics.mean_squared_error(Y_test, y_rbf)
    MSE_lin = metrics.mean_squared_error(Y_test, y_lin)
    MSE_poly = metrics.mean_squared_error(Y_test, y_poly)
    #Apply the model to the validation set
    print("MSE for rbf with K = ", k , "is ", MSE_rbf)
    print("MSE for linear with K = ", k , "is ", MSE_lin)
    print("MSE for poly with K = ", k , "is ", MSE_poly)
    print("\n")
```
